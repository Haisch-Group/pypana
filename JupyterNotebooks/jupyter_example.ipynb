## Step 10 Do sigmoidal fit of data to determine cut-off

```python
from matplotlib import ticker
from scipy.optimize import curve_fit


def sigmoid(x, x0, k):
    y = 1 / (1 + np.exp(-k * (x - x0)))
    return y


n2_popt, n2_pcov = curve_fit(
    sigmoid,
    impactor_n2["n2_sizes"][n2_good_idx],
    impactor_n2["n2_efficiencies"][n2_good_idx],
)
print(n2_popt)
print(f"the cut-off diameter of nozzle 2 is {n2_popt[0]:.2f}" + " \u00b5m")

# create fit function for plotting into data

n2_x = np.linspace(-5, 10, 1000)
n2_y = sigmoid(n2_x, *n2_popt)

# plot it

fig, ax = plt.subplots()

cm = 1 / 2.54  # inches to cm
fig.set_size_inches(16 * cm, 10 * cm)
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())

ax.scatter(
    impactor_n2["n2_sizes"][n2_good_idx],
    impactor_n2["n2_efficiencies"][n2_good_idx] * 100,
    color="#005b7f",
)
ax.plot(n2_x, n2_y * 100, "#005b7f")

plt.xlim(0.07, 12.5)
plt.ylim(0, 100)
ax.set_xscale("log")
ax.set_xticks([0.1, 0.5, 1, 2, 5, 10])
ax.set_xticklabels([0.1, 0.5, 1, 2, 5, 10])
ax.set(xlabel="Particle Diameter / \u00b5m", ylabel="Collection Efficiency %")
ax.legend(["Data", "Fit"])

# save data to png file

from tkinter.filedialog import asksaveasfilename

file_path = asksaveasfilename(
    defaultextension=".png",
    filetypes=[("PNG files", "*.png"), ("JPEG files", "*.jpg"), ("All files", "*.*")],
)
plt.savefig(file_path, dpi=600, transparent=True)

plt.show()
```
